```python
import pandas as pd
```

```python
df=pd.read_csv("/content/titanic.csv")
df.head()
```

```python
inputs=df.drop('Survived',axis='columns')
target=df['Survived']
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_Pclass= LabelEncoder()
le_Sex= LabelEncoder()
le_Age= LabelEncoder()
le_Fare= LabelEncoder()
```

```python
inputs['Pclass_'] = le_Pclass.fit_transform(inputs['Pclass'])
inputs['sex-n']=le_Sex.fit_transform(inputs['Sex'])
inputs['age_']=le_Age.fit_transform(inputs['Age'])
inputs['Fare_']=le_Fare.fit_transform(inputs['Fare'])
inputs
```

```python
input=inputs.drop(['Pclass','Sex','Fare','Ticket','Name','Cabin','Embarked','PassengerId','Age','SibSp','Parch'],axis='columns')
input
```

```python
target
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(input, target)
```

```python
model.score(input, target)
```

```python
model.predict([[1,1,1,7]])
```
